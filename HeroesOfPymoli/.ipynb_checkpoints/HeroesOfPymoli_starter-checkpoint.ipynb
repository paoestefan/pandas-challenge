## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,10, 15, 19, 25, 30,35,39,50]
group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
SN_Clean["Age Demographic"] = pd.cut(SN_Clean["Age"], bins, labels=group_names, include_lowest=False)
SN_Clean
```

```python
 def percent(x):
      pct = float(x/SN_Clean["Age Demographic"].count() * 100)
      return round(pct, 2)
```

```python
A_D=SN_Clean.groupby("Age Demographic")
T_C=A_D["SN"].count()
T_CP=['{:.2f}%'.format(x) for x in T_C.apply(percent)]
T_Cdf=pd.DataFrame({
    "Total Count":T_C,
    "Percentage of Players":T_CP
})
T_Cdf
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order

```python
Top_Spenders=purchase_data.groupby("SN")
TSP_Count=Top_Spenders["Price"].count()
```

```python
TSP_Mean=['${:.2f}'.format(x) for x in Top_Spenders["Price"].mean()]
TSP_Mean
```

```python
TSP_Sum=['${:.2f}'.format(x) for x in Top_Spenders["Price"].sum()]
TSP_Sum
```

```python
TSP_df=pd.DataFrame({
    "Purchase Count":TSP_Count,
    "Average Purchase Price":TSP_Mean,
    "Total Purchase Value":TSP_Sum   
})
TSP_df
```

```python
A_TSP_df=TSP_df.sort_values(by="Total Purchase Value",ascending=False)
A_TSP_df.head()
```
